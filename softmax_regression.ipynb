```python
def show_images(numbers):
    for i in range(1, numbers + 1):
        plt.subplot(5, 10, i)
        image = train.iloc[i].values
        image = image.reshape((28,28))
        plt.imshow(image, cmap='gray')
        plt.xticks([]), plt.yticks([])
        plt.title(y_train[i])

show_images(50)
```

```python
y_train=pd.get_dummies(y_train)
```

```python
y_train
```

```python
from sklearn.model_selection import train_test_split
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(train,y_train,test_size=0.2,random_state=89)
```

```python
import tensorflow as tf
```

```python
x=tf.placeholder(tf.float32,[None,784])
```

```python
w=tf.Variable(tf.zeros([784,10]))
b=tf.Variable(tf.zeros([10]))
```

```python
y=tf.placeholder(tf.float32,[None,10])
```

```python
y_pred=tf.nn.softmax(tf.matmul(x,w)+b)
```

```python
#costfn=-tf.reduce_sum(y*tf.log(y_pred))
costfn=tf.reduce_mean(tf.nn.softmax_cross_entropy_with_logits(logits=y_pred, labels=y))
```

```python
algo=tf.train.GradientDescentOptimizer(0.01).minimize(costfn)
correct_pred=tf.equal(tf.argmax(y_pred,1),tf.argmax(y,1))
accuracy=tf.reduce_mean(tf.cast(correct_pred,tf.float32))
```

```python
init=tf.global_variables_initializer()
session=tf.Session()
session.run(init)
```

```python
for i in range(1000):
    session.run(algo,feed_dict={x:xtrain,y:ytrain})
    print("accuracy="+str(session.run(accuracy,feed_dict={x:xtest,y:ytest})))
```
